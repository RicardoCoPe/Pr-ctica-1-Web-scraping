### Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Activate style alerts
%pycodestyle_on
```

```python
# Loading the data
cars_df = pd.read_csv('cars_dataset.csv')
```

### Numerical analysis and visualization

```python
# Years
cars_df['Year'].plot(kind='box', vert=False)
```

```python
# Years
ax = cars_df['Year'].plot(kind='density')
ax.axvline(cars_df['Year'].mean(), color='red')
ax.axvline(cars_df['Year'].median(), color='green')
```

```python
# Kilometers
cars_df['Kilometers'].plot(kind='box', vert=False)
```

```python
# Kilometers
ax = cars_df['Kilometers'].plot(kind='density')
ax.axvline(cars_df['Kilometers'].mean(), color='red')
ax.axvline(cars_df['Kilometers'].median(), color='green')
```

```python
# Powers
cars_df['Power [CV]'].plot(kind='box', vert=False)
```

```python
# Powers
ax = cars_df['Power [CV]'].plot(kind='density')
ax.axvline(cars_df['Power [CV]'].mean(), color='red')
ax.axvline(cars_df['Power [CV]'].median(), color='green')
```

```python
# Prices
cars_df['Price []'].plot(kind='box', vert=False)
```

```python
# Prices
ax = cars_df['Price []'].plot(kind='density')
ax.axvline(cars_df['Price []'].mean(), color='red')
ax.axvline(cars_df['Price []'].median(), color='green')
```
